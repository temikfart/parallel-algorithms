```python
import numpy as np
from matplotlib import pylab as plt

p = [1, 2, 3, 4, 5, 6, 7, 8]
S_1 = [0.360215, 0.334416, 0.303468, 0.194085, 0.098236, 0.068204, 0.182183, 0.049869]
S_2 = [1.007826, 1.996031, 2.967270, 3.614452, 4.475490, 5.591717, 6.003772, 6.643553]
S_3 = [0.999421, 2.000771, 2.979748, 3.637749, 4.585554, 5.995444, 6.379130, 7.836688]
```

```python
plt.figure(figsize = (15, 10))

plt.plot(p, S_1, color = 'black', label = r'$N = 10^3$')
plt.plot(p, S_2, color = 'yellow', label = r'$N = 10^6$')
plt.plot(p, S_3, color = 'blue', label = r'$N = 10^8$')

plt.title("Ускорение от количества процессов", fontsize = 18)
plt.xlabel("Количество процессов", fontsize = 18)
plt.ylabel("Ускорение", fontsize = 18)

plt.legend(fontsize = 20)
plt.show()
```
