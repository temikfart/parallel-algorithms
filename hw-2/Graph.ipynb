```python
import numpy as np
from matplotlib import pylab as plt
```

```python
p = [1, 2, 3, 4, 5, 6, 7, 8]
```

__a)__ Для $N = 2000$

```python
T1_11 = 102.923319
Tp_11 = np.array([102.923319, 51.035794, 73.989947, 67.824908, 60.858529, 62.475450, 61.885323, 58.344699])
S_11 = T1_11 / Tp_11
```

```python
T1_12 = 102.581700
Tp_12 = np.array([102.581700, 53.061497, 77.171631, 62.529024, 61.711783, 57.211899, 57.893290, 56.269725])
S_12 = T1_12 / Tp_12
```

__b)__ Для $N = 10000$

```python
T1_21 = 12.487523
Tp_21 = np.array([12.487523, 6.130498, 5.234766, 5.255465, 5.813860, 3.188695, 3.026428, 2.822146])
S_21 = T1_21 / Tp_21
```

```python
T1_22 = 12.404149
Tp_22 = np.array([12.404149, 6.118252, 5.225952,  4.413943, 3.643197, 3.225666, 2.936421,  5.474889])
S_22 = T1_22 / Tp_22
```

__b)__ Для $N = 50000$

```python
T1_31 = 15.473638
Tp_31 = np.array([15.473638, 7.711960, 5.751110, 4.284957, 3.550892, 3.076838, 4.699502, 2.341677])
S_31 = T1_31 / Tp_31
```

```python
T1_32 = 15.616074
Tp_32 = np.array([15.616074, 7.830296, 5.697956, 6.293005, 3.589786, 3.005699, 2.597939, 2.310624])
S_32 = T1_32 / Tp_32
```

График

```python
plt.figure(figsize = (15, 10))

plt.plot(p, S_11, color = 'orange', label = 'Fast N = 2000')
plt.plot(p, S_12, linestyle = '--', color = 'orange', label = 'Slow N = 2000')

plt.plot(p, S_21, color = 'red', label = 'Fast N = 10000')
plt.plot(p, S_22, linestyle = '--', color = 'red', label = 'Slow N = 10000')

plt.plot(p, S_31,  color = 'cyan', label = 'Fast N = 50000')
plt.plot(p, S_32, linestyle = '--', color = 'cyan', label = 'Slow N = 50000')

plt.title("Ускорение от количества процессов", fontsize = 18)
plt.xlabel("Количество процессов", fontsize = 18)
plt.ylabel("Ускорение", fontsize = 18)

plt.legend(fontsize = 20)
plt.show()
```
